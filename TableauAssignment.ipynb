Question 1 : What is Tableau? Explain its importance in Business Intelligence and how it helps in data-driven decision-making.
- Tableau is a data visualization tool that helps people see and understand their data. It simplifies raw data into an easily understandable format, enabling users to create interactive dashboards, reports, and charts.
- Its importance lies in empowering users—technical and non-technical—to visually explore data, enabling quick identification of patterns and fostering a data-driven culture for informed, real-time decisions across industries like finance, retail, and healthcare.

Question 2 : Explain the role of the following Tableau components:
a) Data Pane
b) Worksheet
c) Dashboard
d) Story
- Data Pane: The Data Pane is located on the side of a worksheet and displays all the fields (dimensions and measures) available from the connected data source
- Worksheet: A Worksheet is the fundamental canvas where individual data visualizations, such as a single chart or a graph, are created
- Dashboard: A Dashboard is a consolidated view that brings together multiple worksheets and related information in a single screen
- Story: A Story contains a sequence of worksheets or dashboards (called story points) that work together in a guided narrative

Question 3 : What is the difference between Dimensions and Measures in Tableau? Provide examples of each.
- Dimensions are qualitative, descriptive data that segment your data and answer "what/who/where/when," appearing as blue discrete headers.
- Measures are quantitative, numerical data that you aggregate to answer "how much/many," appearing as green continuous pills, creating axes. Dimensions provide context and detail, while measures provide values for calculations and visualizations.

Question 4 : Define and explain the purpose of Filters, Parameters, and Sets in Tableau.
- Filters reduce data shown in a view  
- Parameters provide user-controlled, single-value inputs
- Sets group data members for comparisons, all enhancing interactivity and analysis by letting users dynamically slice, control, and compare data without editing the underlying data source.
